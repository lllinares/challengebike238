```python
def create_geojson_features(data):
    features = []
    
    for row in data:
        feature = {
            'type': 'Feature',
            'geometry': {
                'type':'Point', 
                'coordinates':[row['location']['coordinates'][0],row['location']['coordinates'][1]]},
            'properties': {
                'time': pd.to_datetime(row['dateObserved'][:10]).__str__(),
                'style': {'color' : ''},
                'icon': 'circle',
                'iconstyle':{
                    'fillColor': '#53c688',
                    'fillOpacity': 0.8,
                    'stroke': 'true',
                    'radius': row['intensity']/75
                }
            }
    }
        features.append(feature)
    return features

start_geojson = create_geojson_features(table)
```

```python
from folium.plugins import TimestampedGeoJson


nyc_map = folium.Map(location = [ 43.61465,3.8336],
                    tiles = "CartoDB Positron",
                    zoom_start = 14)


TimestampedGeoJson(start_geojson,
                  ).add_to(nyc_map)
nyc_map.save("montmap.html")
```
